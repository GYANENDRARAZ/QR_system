```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk, ImageDraw, ImageFont
import qrcode

# ====================== FUNCTIONS ======================

# Convert Vehicle Number to CAPITAL
def convert_uppercase(event):
    vehicle_text.set(vehicle_text.get().upper())


# ====================== GENERATE QR ======================

def generate_qr():

    name = name_entry.get()
    mobile = mobile_entry.get()
    vehicle = vehicle_entry.get()

    # ================= VALIDATION =================

    if name == "" or mobile == "" or vehicle == "":
        messagebox.showerror("Error", "Please fill all details")
        return

    if not mobile.isdigit():
        messagebox.showerror(
            "Invalid Mobile Number",
            "Mobile number should contain only digits"
        )
        return

    if len(mobile) != 10:
        messagebox.showerror(
            "Invalid Mobile Number",
            "Mobile number must be exactly 10 digits"
        )
        return

    # ================= QR DATA =================

    qr_data = f"""
Name : {name}
Mobile : {mobile}
Vehicle No : {vehicle}
"""

    # ================= GENERATE QR =================

    qr = qrcode.make(qr_data)

    qr_file = "temp_qr.png"

    qr.save(qr_file)

    # ================= CREATE CARD =================

    card = Image.new("RGB", (900, 600), "#eef2f7")

    draw = ImageDraw.Draw(card)

    # ================= FONTS =================

    title_font = ImageFont.truetype("arial.ttf", 42)

    label_font = ImageFont.truetype("arial.ttf", 24)

    value_font = ImageFont.truetype("arial.ttf", 38)

    scan_font = ImageFont.truetype("arial.ttf", 24)

    footer_font = ImageFont.truetype("arial.ttf", 18)

    # ================= HEADER =================

    draw.rectangle((0, 0, 900, 120), fill="#0f2c7d")

    # CENTER TITLE
    draw.text(
        (170, 30),
        "VEHICLE_QR_ID_CARD",
        fill="white",
        font=title_font
    )

    # Bold Effect
    draw.text(
        (171, 30),
        "VEHICLE_QR_ID_CARD",
        fill="white",
        font=title_font
    )

    # ================= MAIN CARD =================

    draw.rounded_rectangle(
        (45, 165, 855, 545),
        radius=30,
        fill="white",
        outline="#cbd5e1",
        width=5
    )

    # ================= OWNER NAME =================

    draw.text(
        (90, 190),
        "OWNER NAME",
        fill="#42526e",
        font=label_font
    )

    draw.text(
        (90, 225),
        name.upper(),
        fill="black",
        font=value_font
    )

    # Bold Effect
    draw.text(
        (91, 225),
        name.upper(),
        fill="black",
        font=value_font
    )

    # ================= MOBILE NUMBER =================

    draw.text(
        (90, 330),
        "MOBILE NUMBER",
        fill="#42526e",
        font=label_font
    )

    draw.text(
        (90, 350),
        mobile,
        fill="black",
        font=value_font
    )

    draw.text(
        (91, 350),
        mobile,
        fill="black",
        font=value_font
    )

    # ================= VEHICLE NUMBER =================

    draw.text(
        (89, 460),
        "VEHICLE NUMBER",
        fill="#42526e",
        font=label_font
    )

    draw.text(
        (90, 480),
        vehicle.upper(),
        fill="black",
        font=value_font
    )

    draw.text(
        (91, 480),
        vehicle.upper(),
        fill="black",
        font=value_font
    )

    # ================= QR BOX =================

    draw.rounded_rectangle(
        (560, 220, 800, 500),
        radius=25,
        fill="white",
        outline="#d6deea",
        width=4
    )

    # ================= QR IMAGE =================

    qr_img = Image.open(qr_file)

    qr_img = qr_img.resize((180, 180))

    card.paste(qr_img, (590, 250))

    # ================= SCAN TEXT =================

    draw.text(
        (630, 455),
        "SCAN ME",
        fill="#1450e0",
        font=scan_font
    )

    # ================= FOOTER =================

    draw.text(
        (280, 565),
        "Generated by Gyanendra Raz",
        fill="#64748b",
        font=footer_font
    )

    # ================= SAVE CARD =================

    final_file = f"{vehicle}_card.png"

    card.save(final_file)

    # ================= SHOW QR IN GUI =================

    display_qr = qr_img.resize((180, 180))

    photo = ImageTk.PhotoImage(display_qr)

    qr_label.config(image=photo)

    qr_label.image = photo

    result_label.config(
        text=f"Card Saved Successfully\n{final_file}"
    )

    messagebox.showinfo(
        "Success",
        f"Vehicle Card Saved Successfully\n{final_file}"
    )


# ====================== MAIN WINDOW ======================

root = tk.Tk()

root.title("Vehicle QR Registration")

root.geometry("820x680")

root.config(bg="#eaf2ff")

root.resizable(False, False)

# ====================== TITLE ======================

title = tk.Label(
    root,
    text="🚗 Vehicle QR Registration Form",
    font=("Segoe UI", 24, "bold"),
    bg="#eaf2ff",
    fg="#0f172a"
)

title.pack(pady=20)

# ====================== MAIN FRAME ======================

main_frame = tk.Frame(
    root,
    bg="white",
    bd=2,
    relief="ridge"
)

main_frame.place(x=70, y=90, width=680, height=520)

# ====================== FORM FRAME ======================

form_frame = tk.Frame(
    main_frame,
    bg="white"
)

form_frame.place(x=25, y=20)

heading = tk.Label(
    form_frame,
    text="Registration Details",
    font=("Segoe UI", 18, "bold"),
    bg="white",
    fg="#1e293b"
)

heading.grid(row=0, column=0, columnspan=2, pady=15)

# ====================== NAME ======================

name_label = tk.Label(
    form_frame,
    text="Full Name",
    font=("Segoe UI", 12, "bold"),
    bg="white"
)

name_label.grid(row=1, column=0, sticky="w", pady=10)

name_entry = tk.Entry(
    form_frame,
    font=("Segoe UI", 12),
    width=30,
    bd=2,
    relief="groove"
)

name_entry.grid(row=1, column=1, pady=10, padx=10, ipady=5)

# ====================== MOBILE ======================

mobile_label = tk.Label(
    form_frame,
    text="Mobile Number",
    font=("Segoe UI", 12, "bold"),
    bg="white"
)

mobile_label.grid(row=2, column=0, sticky="w", pady=10)

mobile_entry = tk.Entry(
    form_frame,
    font=("Segoe UI", 12),
    width=30,
    bd=2,
    relief="groove"
)

mobile_entry.grid(row=2, column=1, pady=10, padx=10, ipady=5)

# ====================== VEHICLE ======================

vehicle_label = tk.Label(
    form_frame,
    text="Vehicle Number",
    font=("Segoe UI", 12, "bold"),
    bg="white"
)

vehicle_label.grid(row=3, column=0, sticky="w", pady=10)

vehicle_text = tk.StringVar()

vehicle_entry = tk.Entry(
    form_frame,
    textvariable=vehicle_text,
    font=("Segoe UI", 12),
    width=30,
    bd=2,
    relief="groove"
)

vehicle_entry.grid(row=3, column=1, pady=10, padx=10, ipady=5)

vehicle_entry.bind("<KeyRelease>", convert_uppercase)

# ====================== BUTTON ======================

generate_btn = tk.Button(
    form_frame,
    text="Generate QR",
    font=("Segoe UI", 11, "bold"),
    bg="#1450e0",
    fg="white",
    width=18,
    height=1,
    bd=0,
    cursor="hand2",
    command=generate_qr
)

generate_btn.grid(row=4, column=0, columnspan=2, pady=25)

# ====================== RESULT ======================

result_label = tk.Label(
    form_frame,
    text="",
    font=("Segoe UI", 10),
    bg="white",
    fg="green"
)

result_label.grid(row=5, column=0, columnspan=2)

# ====================== QR FRAME ======================

qr_frame = tk.Frame(
    main_frame,
    bg="#f8fafc",
    bd=2,
    relief="ridge"
)

qr_frame.place(x=430, y=60, width=210, height=320)

qr_title = tk.Label(
    qr_frame,
    text="Generated QR",
    font=("Segoe UI", 15, "bold"),
    bg="#f8fafc",
    fg="#0f172a"
)

qr_title.pack(pady=15)

qr_label = tk.Label(
    qr_frame,
    bg="#f8fafc"
)

qr_label.pack(pady=10)

# ====================== FOOTER ======================

footer = tk.Label(
    root,
    text="Developed by Gyanendra Raz",
    font=("Segoe UI", 10),
    bg="#eaf2ff",
    fg="#475569"
)

footer.pack(side="bottom", pady=10)

# ====================== RUN ======================

root.mainloop()
```
